```python
name = "Python for Beginners"
# index 012345678910
print(name)

print(name[0:5:2])   # Skipping in whole String.
print(name[0::2])  # Skipping in whole String 1 index


# # Reverse String
print(name[::-1])
```

```python
name = "Python for Beginners"
# index  012345678910
print(name)

# Reverse Slicing
print(name[:-2])

print(name[1::-2])
```

## Itertools

```python
import itertools
   
# for in loop
for i in itertools.count(5, 5):
    if i == 35:
        break
    else:
        print(i, end =" ")
```

```python
import itertools
count = 0
# for in loop
for i in itertools.cycle('AB'):
    if count > 7:
        break
    else:
        print(i, end = " ")
        count += 1
```

```python
import itertools
l = ['Geeks', 'for', 'Geeks']
# defining iterator
iterators = itertools.cycle(l)
# for in loop
for i in range(6):
    # Using next function
    print(next(iterators), end = " ")
```
