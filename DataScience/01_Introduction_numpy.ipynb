## 1. 🧠 What is NumPy?

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.


At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types

## 2. ⚖️ NumPy Arrays vs Python Sequences

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

**Quick example — the #1 beginner trap:** `*` means something completely different for a list vs an array.

```python
import numpy as np

list_a = [1, 2, 3]
arr_a = np.array([1, 2, 3])

print(list_a * 2)   # [1, 2, 3, 1, 2, 3]  -> repeats the list
print(arr_a * 2)    # [2 4 6]             -> multiplies every element
```

**More examples — the fixed-size & same-dtype rules in action:**

```python
arr = np.array([1, 2, 3])
try:
    arr.append(4)   # lists have .append(), arrays don't
except AttributeError as e:
    print("Error:", e)
```

```python
# mixing types: NumPy picks ONE dtype for everything (upcasting)
mixed = np.array([1, 2, 3.5])          # ints + float -> all become float
print(mixed, mixed.dtype)

mixed_str = np.array([1, 2, "three"])  # any string present -> all become strings
print(mixed_str, mixed_str.dtype)
```

## 4. 📐 Array Attributes

**Attributes vs methods:** `.shape`, `.ndim`, `.size`, `.dtype`, `.itemsize` are *attributes* — no `()` at the end, they just describe the array. Things like `.reshape()` or `.astype()` are *methods* that do work, and need `()`.

| Attribute | Tells you |
|---|---|
| `.ndim` | how many dimensions |
| `.shape` | size along each dimension |
| `.size` | total number of elements |
| `.dtype` | data type of the elements |
| `.itemsize` | bytes used per element |

```python
a1 = np.arange(10,dtype=np.int32)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)

a3
```

```python
# ndim
print(a1.ndim) # 1D Array
print(a2.ndim) # 2D Array
print(a3.ndim) # 3D Array
```

```python
# shape
print(a3.shape)
a3
```

```python
# shape
print(a2.shape)
a2
```

```python
# size
print(a2.size)
a2
```

```python
# itemsize
print(a1.itemsize) # 32bit int
print(a3.itemsize) # 64bit int
```

Smaller dtypes save memory but hold a smaller range of numbers. `int32` (4 bytes) holds roughly ±2 billion; `int64` (8 bytes) holds a vastly bigger range but uses twice the memory per number.

```python
# dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)

```

**More attribute examples:**

```python
# .nbytes = total memory used by the WHOLE array (itemsize * size)
print(a2.itemsize, a2.size, a2.itemsize * a2.size)
print(a2.nbytes)
```

```python
# attributes update automatically after reshaping -- they describe the CURRENT shape
b = np.arange(24)
print(b.shape, b.ndim)
b = b.reshape(2,3,4)
print(b.shape, b.ndim)
```

## 6. ➕ Array Operations

Two kinds of operations: **scalar** (array + a single number, e.g. `a * 2`) applies the same way to every element; **vector** (array + array) works *element by element*, and normally needs matching shapes — unless "broadcasting" kicks in (see below).

| Type | Example | What happens |
|---|---|---|
| Scalar | `a * 2` | every element ×2 |
| Scalar | `a ** 2` | every element squared |
| Vector | `a1 + a2` | element-by-element add (same shape) |
| Relational | `a > 15` | element-by-element compare → boolean mask |

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)

a2
```

```python
a1
```

```python
a1 * 2
```

```python
# scalar operations
print(a1)
# arithmetic
a1 ** 2 ## square
```

```python
# relational
print(a2)
a2 == 15
```

```python
print(a2)
a2 > 15
```

Comparisons like `==`, `>`, `<` don't return a single True/False — they return a whole array of True/False values, one per element. This is called a **boolean mask**, and it's the key to filtering data (see *Boolean & Fancy Indexing* further down).

```python
# vector operations
# arithmetic
a1 ** a2
```

**Gotcha — integer overflow:** notice some results above are enormous, and one is even negative! `int64` can only hold values up to about 9.2×10¹⁸ — once `a1 ** a2` goes past that, the value silently wraps around into nonsense/negative territory instead of raising an error. If results can get this large, use a `float` dtype instead.

```python
print(a1 + a2)
a1 + a2
```

**Broadcasting:** NumPy can still do vector operations even when array shapes don't match exactly, by mentally "stretching" the smaller one to fit — no need to manually copy it into the same shape yourself.

```python
row = np.arange(4)                  # shape (4,)
grid = np.arange(12).reshape(3, 4)  # shape (3,4)

print(row)
print(grid + row)   # row is broadcast: added to EVERY row of grid
print(grid + 10)     # a single number is broadcast to every element too
```

**More operation examples:**

```python
# division: NumPy follows true division by default (like Python 3's /)
print(a1 / 2)     # true division -> floats
print(a1 // 2)    # floor division -> whole numbers
```

```python
# chain a comparison mask straight into indexing to filter values
print(a1[a1 % 3 == 0])   # every element in a1 that's divisible by 3
```

## 7. 📊 Array Mathematical Functions

These functions summarize the whole array into fewer numbers (or just one). Most accept an `axis` argument so you can summarize *per row* or *per column* instead of the entire array — see the mnemonic below.

| Function | What it gives you |
|---|---|
| `np.max` / `np.min` | largest / smallest value |
| `np.sum` / `np.prod` | total sum / total product |
| `np.mean` / `np.median` | average / middle value |
| `np.std` / `np.var` | spread of the values |

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
np.max(a1)
```

```python
np.min(a1)
```

```python
np.sum(a1)
```

```python
np.prod(a1)
```

```python
# max/min/sum/prod
# 0 -> col and 1 -> row
np.max(a1,axis=0) ## column vise
```

```python
# max/min/sum/prod
# 0 -> col and 1 -> row
np.max(a1,axis=1) ## row wise (one result per row)
```

```python
# max/min/sum/prod
# 0 -> col and 1 -> row
np.prod(a1,axis=0)
```

**Axis mnemonic:** `axis=0` walks *down* the rows, collapsing them → one result **per column**. `axis=1` walks *across* the columns, collapsing them → one result **per row**. If unsure, check the output shape — it tells you which axis disappeared.

**More math function examples:**

```python
# cumsum/cumprod: running totals, one step at a time
b = np.array([1,2,3,4])
print(np.cumsum(b))    # [1, 3, 6, 10]
print(np.cumprod(b))   # [1, 2, 6, 24]
```

### 9.2 🔍 Boolean & Fancy Indexing
Two more powerful ways to pick elements, beyond plain slicing:
- **Boolean indexing** — pass a True/False mask (like the one from `a2 > 15` earlier) to keep only the matching elements.
- **Fancy indexing** — pass a list of indices to pick out several specific elements at once, in any order.

```python
nums = np.arange(20)

# Boolean indexing: keep only values greater than 12
mask = nums > 12
print(mask)         # array of True/False
print(nums[mask])   # only the True positions survive

# Fancy indexing: grab specific positions directly
print(nums[[0, 5, 19]])   # elements at index 0, 5, and 19
```

**`np.where()` — get positions, not just values:**

```python
nums = np.arange(20)
idx = np.where(nums > 12)
print(idx)          # a tuple of index arrays where the condition is True
print(nums[idx])    # using those indices to fetch the values (same as nums[nums>12])
```

```python
# np.where can also pick between two values based on a condition
np.where(nums % 2 == 0, "even", "odd")
```

## 11. 🧩 Reshaping

Reshaping rearranges the *same* data into a new shape (nothing is added, removed, or changed) — it usually returns a *view* of the original array rather than a copy, so editing one can sometimes affect the other.

```python
# reshape
a2
```

```python
# Transpose
np.transpose(a2)
a2.T
```

```python
# ravel
a3.ravel() #convert 3D and other D array to 1D array
```

**`ravel()` vs `flatten()`:** both turn any-shaped array into 1D, but `ravel()` returns a *view* (shares memory with the original — editing one can edit the other) while `flatten()` always returns an independent *copy*. When unsure, `flatten()` is the safer default.

```python
grid = np.arange(6).reshape(2, 3)
view = grid.ravel()
view[0] = 99
print(grid)   # original changed too! ravel() shared memory

copy = grid.flatten()
copy[0] = -1
print(grid)   # original untouched -- flatten() made a real copy
```

**Reshape shortcuts & resize:**

```python
# use -1 in reshape to let NumPy figure out that dimension automatically
b = np.arange(12)
print(b.reshape(3, -1))   # NumPy computes -1 -> 4
print(b.reshape(-1, 6))   # NumPy computes -1 -> 2
```

```python
# reshape() refuses to change the total element count; resize() will (repeats data)
try:
    b.reshape(5, 5)
except ValueError as e:
    print("reshape error:", e)

print(np.resize(b, (4, 4)))   # 16 slots needed from 12 elements -> wraps around
```

## 12. 🔗 Stacking

Mnemonic: **h**stack = **h**orizontal, side-by-side (rows must match). **v**stack = **v**ertical, stacked on top of each other (columns must match).

```python
# horizontal stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
a5
```

```python
a4
```

```python
## Horizontak stacking
np.hstack((a4,a5))
```

```python
# Vertical stacking
np.vstack((a4,a5))
```

**`np.concatenate()` — the general form of stacking:**

```python
# hstack/vstack are shortcuts for concatenate along a specific axis
print(np.concatenate((a4, a5), axis=1))   # same result as hstack
print(np.concatenate((a4, a5), axis=0))   # same result as vstack
```

```python
# stack multiple arrays into a brand-new extra dimension
np.stack((a4, a5))   # shape becomes (2, 3, 4) -- a "batch" of two (3,4) arrays
```

## 13. ✂️ Splitting

Splitting is the reverse of stacking — it cuts one array into several equal-sized pieces. The split count must divide evenly into that axis's length (e.g. `hsplit(a4, 4)` needs exactly 4 columns).

```python
# horizontal splitting
a4
```

```python
np.hsplit(a4,4)
```

```python
# vertical splitting
a5
```

```python
np.vsplit(a5,3)
```

**`np.array_split()` — splitting into uneven pieces:**

```python
# hsplit/vsplit need the split count to divide evenly -- array_split doesn't
b = np.arange(10)
print(np.array_split(b, 3))   # 10 doesn't divide evenly into 3 -- extras go first
```

```python
# split at specific custom positions instead of evenly
print(np.split(b, [2, 5]))   # cut right before index 2, and right before index 5
```

## 14. ✅ Key Takeaways
- NumPy arrays are fixed-size, single-dtype grids — that's exactly what makes them fast.
- `*` and other operators act *element-wise* on arrays, unlike on Python lists.
- Attributes (`.shape`, `.ndim`, `.size`) describe an array; methods (`.reshape()`, `.astype()`) transform it.
- `axis=0` collapses down the rows (per-column result); `axis=1` collapses across the columns (per-row result).
- Indexing/slicing order is always left-to-right, outermost dimension first: `[row, col]` for 2D, `[depth, row, col]` for 3D.
- Boolean masks (`arr > 5`) and fancy indexing (`arr[[1,3,5]]`) are the main ways to filter/select data.
- Broadcasting lets NumPy combine arrays of different (but compatible) shapes without you manually copying data.
- Watch for silent gotchas: `.astype()` truncates instead of rounding, and very large integer operations can silently overflow.
