# List In python

#### lists allow us to store multiple items in a single variable. For example, if you need to store the ages of all the students in a class, you can do this task using a list. Lists are similar to arrays (dynamic arrays that allow us to store items of different data types) in other programming languages.

- Ordered - They maintain the order of elements.
- Mutable - Items can be changed after creation.
- Allow duplicates - They can contain duplicate values.

```python
names = ["Ravi", "Rahul", "Praneet"]
print(names[1])
```

# List Operations in Python Language

```python
## len() - for finding the length of the list

myList = [2,3,4, "Word", "String"]
print("Length of the list is: ",len(myList))
```

```python
## Negative Indexing

number = [12, 3, 90, 2, 4, 5, 6, 7]
print(number[-2]) ## print the number from reverse list


#negative indexing with lists
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
```

```python
## append() in List
num = []
num.append(199)  # Add element in last
print(num)

x = [1,2,3]
x.append(4)
print(x)
sum(x)
print(sum(x))
```

```python
myList = [2,3,4, "Word", "String"]
myList.append("Append")
print(myList)

myList.insert(4, "Four") #insert Using Index
print(myList)

myList.reverse() #Reverse of list
print(myList)


#adding elements to an empty list
l = []

for i in range(10):
    l.append(i)
l
```

```python
## extend() in the list with any data

number = [12, 3, 9, 2, 4, 5, 6, 7]
myList = [12, 3, 9, 2, 4, 5, 6, 7]
print(number)
number.append(90)

number.append(["new1", "new2", "new3"])
print(number)

myList.extend(["new1", "new2", "new3"]) # basically used for extending the list in the current list
print(myList)
```

```python
## insert() - to insert the data in the list in any place using indexing

myList = [12, 3, 9, 2, 4, 5, 6, 7]

print(myList)
myList.insert(2, 100)
print(myList)
```

```python
## pop() and remove() function in python

numr = [1, 2, 3, 4, 6]
numr.remove(1)  # Remove element from list
print(numr)


numPop = [1, 2, 3, 4, 6]
numPop.pop(1)  # Pop element from list by index
print(numPop)

```

```python
## clear() - is used to clear the list

myList = [12, 3, 9, 2, 4, 5, 6, 7]

print(myList)

myList.clear()

print(myList)
```

```python
## min() - is used to find the minimum value

myList = [12, 3, 9, 2, 4, 5, 6, 7]
print(min(myList)) # O(N)

## max() - is used to find the maximum value
myList = [12, 3, 9, 2, 4, 5, 6, 7]
print(max(myList)) # O(N)
```

```python
## count() is used to count the occurence of element in the list

myList = [12, 3, 9, 2, 2, 5, 6, 2]

print(myList.count(2)) # find the occurence of the number 

myList = [2,4,4,4,1,7,3,8,2,9,3,2,9,2,4,6,8,2,3,9,2]

myList.sort() ## sort() method used to sort the data in the list - O(nLogN)
print(myList)

myList.reverse() ##reverse() method is used to reverse the data in the list - O(N)
print(myList)

```

```python
# index() is used to find the index of the data in list

myList = [12, 3, 9, 2, 2, 5, 6, 2]

# print(myList.index(0))
print(myList.index(12)) 
```

```python
# append() function in python
# copy function in python

list1 = [1, 2, 3]
list2 = list1 # deep copy
list3 = list1.copy() # shallow copy
list2.append(4)

print(list1)
print(list2)
print(list3)
```

## List Comprehensions

 List comprehensions are used for creating new lists from other iterables.

As list comprehensions return lists, they consist of brackets containing the expression, which is executed for each element along with the for loop to iterate over each element.

```python
#example:
numbers = [1, 2, 3, 4]
squares = [n**2 for n in numbers]

print(squares)
```

* append()	Adds an element at the end of the list
* clear()	Removes all the elements from the list
* copy()	Returns a copy of the list
* count()	Returns the number of elements with the specified value
* extend()	Add the elements of a list (or any iterable), to the end of the current list
* index()	Returns the index of the first element with the specified value
* insert()	Adds an element at the specified position
* pop()	Removes the element at the specified position
* remove()	Removes the item with the specified value
* reverse()	Reverses the order of the list
* sort()	Sorts the list
