## Oops In Python

- Class is a blueprint for creating objects


```python
class Students:
    college_name = "ABC University"
    name = "Karan" # class Attribute
    def __init__(self, name, marks):
       self.name = name # Object Attribute
       self.marks = marks
       print("Adding in Databases")
       
    def welcome(self):
        print("welcome student,", self.name)
        
    def get_marks(self):
        return self.marks
        
    
    

s1 = Students("Pratham", 234)
print(s1.name, s1.marks)

s2 = Students("Rahul", 345)
print(s1.name, s1.marks)

print(s2.college_name)

s1.welcome()

print(s1.get_marks())
```

```python
## Lets Practice

class Student:
    
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        
    def get_avg(self):
        sum = 0
        for val in self.marks:
            sum += val
            
        print("Hi", self.name, "You avg score is: ", sum/3)
        
    @staticmethod ## without self this is used wth the def methods
    def hello_print():
        print("Hello Pratham")
        
        
s1 = Student("Tony Stark", [33, 45, 67])

s1.get_avg()

s1.name = "Iron Man"

s1.get_avg()


s1.hello_print()
```

## Abstraction
- Hiding the implementation details of a class and only showing the essentials feature to the user

```python
class Car:
    
    def __init__(self):
        self.acc = False
        self.brk = False
        self.clutch = True
    
    def start(self):
        self.clutch = False
        self.acc = True
        print("Car Started...")
        
    
    def stop(self):
        self.brk = True
        self.clutch = True
        print("Car Stopped...")
    
toyata = Car()

toyata.start()
toyata.stop()
    
```

## Encapsulation
- Wrapping data and functions into a single unit(Object)
 
### Lets practice 
- create account class with 2 attributes - balance and account no 
- create methods for debit, credit and printing the balance

```python
class Accounts:
    
    def __init__(self, bal, acc):
        self.bal = bal
        self.account_no = acc
        
    def debit_method(self, amt):
        self.bal -= amt
        print("Rs. ", amt, "was debited")   
        print("Total Balance = ", self.get_balance()) 
        
    def credit_method(self, amt):    
        self.bal += amt
        print("Rs. ", amt, "was credited") 
        print("Total Balance = ", self.get_balance())    
        
    def get_balance(self):
        return self.bal 
    
    
    
    
acc1 = Accounts(10000, "Pratham")
acc1.debit_method(1250)
acc1.credit_method(250)
acc1.credit_method(40000)
acc1.debit_method(250)  
```
