# Functions

```python
#Defining Function
def addTwoNumber(a,b):
  c = a+b
  print(c)

#Calling Function
addTwoNumber(2,3)
```

```python
#defining a function
def my_function():
  print("Hello This is a function")

my_function()
```

```python
#passing arguments
def my_function(fname):
  print(fname + " Kumar")   #always call function with the require number of args;not less,not more


my_function("Pratham")
```

```python
#calling
my_function("Shubham")
```

```python
ig = ["Worry", "Reast", "Rohan", "Tejas"]

#arbitrary arguments: if number of arguments is unknown, add a * before the arg
def my_function(*kids):
  print("The youngest child is " + kids[1])

my_function("Worry", "Reast")
```

```python
#keyword arguments:You can also send arguments with the key = value syntax.This way the order of the arguments does not matter.
def my_function(child3, child2, child1):
  print("The youngest child is " + child3)
  print("The youngest child is " + child2)

my_function(child1 = "Emil", child2 = "Tobias", child3 = "Linus")
```

```python
#arbitrary keyword arguments, **kwargs
def my_function(**kid):
  print("His last name is " + kid["lname"])

my_function(fname = "Tobias", lname = "Refsnes")
```

```python
#default parameters
def my_function(country = "Norway"):
  print("I am from " + country)

my_function()
```

```python
#return values
def my_function(x):
  return 5 * x


my_function(4)
```
