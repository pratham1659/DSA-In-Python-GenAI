## Remove Html Tags Operations

```python
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

## Remove Url Operation

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

## Punctuation Handling Operation

```python
import string,time
string.punctuation
```

```python
exclude = string.punctuation
exclude
```

```python
def remove_punc(text):
    for char in exclude:  ## linear time complexity to remove the punctuations
        text = text.replace(char,'')
    return text
```

```python
text = 'string. With. Punctuation?'
```

```python
start = time.time()  ## Example to remove the punctuations
print(remove_punc(text))
time1 = time.time() - start
print(time1*50000)
```

```python
def remove_punc1(text): ## another method to remove the punctuations
    return text.translate(str.maketrans('', '', exclude)) ## Complexity of O(N)
```

```python
start = time.time()
remove_punc1(text)
time2 = time.time() - start
print(time2*50000)
```

```python
time1/time2
```

## Chat Conversion Handle Operations

```python
chat_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As Soon As Possible',
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "BRB": "Be Right Back",
    "BTW": "By The Way",
    "OMG": "Oh My God",
    "IMO": "In My Opinion",
    "LOL": "Laugh Out Loud",
    "TTYL": "Talk To You Later",
    "GTG": "Got To Go",
    "TTYT": "Talk To You Tomorrow",
    "IDK": "I Don't Know",
    "TMI": "Too Much Information",
    "IMHO": "In My Humble Opinion",
    "ICYMI": "In Case You Missed It",
    "AFAIK": "As Far As I Know",
    "BTW": "By The Way",
    "FAQ": "Frequently Asked Questions",
    "TGIF": "Thank God It's Friday",
    "FYA": "For Your Action",
    "ICYMI": "In Case You Missed It",
}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
chat_conversion('Do this work ASAP')
```

```python
# 1. Non-capturing pattern (silences warning in .str.contains)
check_pattern = r"\b(?:" + "|".join(map(re.escape, chat_words.keys())) + r")\b"

# 2. Capturing pattern (required for .str.extract)
extract_pattern = (
    r"\b(" + "|".join(map(re.escape, chat_words.keys())) + r")\b"
)

# 3. Filter rows quietly using the check_pattern
matched_df = df[
    df["review"].str.contains(check_pattern, case=False, na=False)
].copy()

# 4. Create the 3rd column by extracting the exact word that matched
matched_df["found_word"] = matched_df["review"].str.extract(
    extract_pattern, flags=re.IGNORECASE
)

# 5. Display the neat 3-column table
matched_df[["found_word", "review"]]
```

```python
df["review"] = df["review"].apply(chat_conversion)
```

## Stopwords Operations

```python
from nltk.corpus import stopwords ## NLTK: Natural Level ToolKit library
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english') ## Identify the language and pass here
```

```python
len(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
remove_stopwords('probably my all-time favorite movie, a story of selflessness, sacrifice and dedication to a noble cause, but it\'s not preachy or boring. it just never gets old, despite my having seen it some 15 or more times')
```

## Remove Emoji Handle Operations

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emoji("Loved the movie. It was 😘😘")
```

# Stemmer

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

# Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations="?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
