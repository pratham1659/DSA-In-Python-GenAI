# Practive problems in Loops and Control Statements

```python
# FIZZBUZZ Problem | LEETCODE - 412

from typing import List


def fizzbuzz(n: int) -> List[str]:
    ans = []
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            ans.append(str(i))
    return ans
            
                   
print(fizzbuzz(10), end=" ") 
```

```python
## Count Odd Numbers in an Interval Range | LEETCODE - 1523
low = 3
high = 7

def countOdds(low: int, high: int) -> int:
    return ((high + 1) // 2) - (low // 2)


print(countOdds(3, 7))

```

```python
## How Many Numbers Are Smaller Than the Current Number | LEETCODE - 1365

from typing import List

def smallerNumberCount(nums: List[int]) -> List[int]:
    ans = []
    for i in nums:
        count = 0
        for j in nums:
            if j < i:
                count += 1
        ans.append(count)
    return ans
    
    
nums = [8, 1, 2, 2, 3]    
print(smallerNumberCount(nums))
```

```python
## Find the count of number using while loop

num = 5183
count = 0
sum = 0
while num>0:
    r = num%10
    num //= 10
    sum = r + sum
    count+=1
    

print("Count of Digits", count)
print("Sum of digits", sum)

```

```python
## Count the Digits That Divide a Number | LEETCODE - 2520

def countDigits(num: int) -> int:
    temp = num
    ans = 0
    while temp > 0:
        r = temp % 10
        if num % r == 0:
            ans+=1
        temp //= 10
    return ans
        
num = 121 
print(countDigits(num))
```

```python
## Palindrome Number | LEETCODE - 9

def checkPalindrome(num: int) -> bool:
    ans = 0
    temp = num
    while temp > 0:
        r = temp % 10
        ans = ans * 10 + r
        temp //=10
    
    return num == ans
        

num = 212
print(checkPalindrome(num))
```

```python
## Subtract the Product and Sum of Digits of an Integer | LEETCODE - 1281

def subtractProductAndSum(num: int) -> int:
    temp = num
    prod = 1
    add = 0
    
    while temp > 0:
        r = temp % 10
        temp //= 10
        add+= r
        prod*= r
    
    return prod - add
        
    
num = 234
print(subtractProductAndSum(num))
```

```python
## Kids With the Greatest Number of Candies | LEETCODE - 1431
from typing import List

def kidsWithCandies(candies: List[int], extraCandies: int) -> int:
    maxCandies = max(candies)
    ans = []
    
    for i in candies:
        if (i+extraCandies) >= maxCandies:
            ans.append(True)
        else:
            ans.append(False) 
    return ans   

candies = [2,3,5,1,3]
extraCandies = 3   
print(kidsWithCandies(candies, extraCandies))
    
```
