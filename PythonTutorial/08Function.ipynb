## Recursion Function

- The process in which a function calls itself directly or indirectly is called recursion and the corresponding function is called a recursive function.

### Loops vs Recursion


```python
n = 10
i = 1

while i<=n:
    print(i, end=" ")
    i+=1
```

```python
def printNumbers(i, n):
    # base case
    if i > n:
        return
    
    # recursive case
    print(i, end=" ")
    printNumbers(i+1, n)
    
printNumbers(1, 10)
```

```python
def factorialNum(n):
    if n==0:
        return 1
    return n*factorialNum(n-1)

print(factorialNum(5))
    
```

```python
## Recursive Stack
def fun(n):
    if n==0:
        return 
    fun(n - 1)
    print(n, end=" ")

fun(3)
    
```

```python
##Fibonacci Series
# - F(0) = 0, F(1) = 1
# F(n) = F(n - 1) + F(n - 2), for n > 1.

def fibo(n):
    # base case
    if n == 0 or n == 1:
        return n
    
    # recursive case
    return fibo(n-1) + fibo(n-2)
    
print(fibo(6))
```

```python
## Practice Problem | 1137. N-th Tribonacci Number

def trifibo(n):
    #base case
    if n == 0 or n == 1:
        return n
    if n == 2:
        return 1 # Or 2, depending on your problem's specific starting sequence
    #recursive case
    return trifibo(n-1) + trifibo(n-2) + trifibo(n-3)

print(trifibo(25))
    
```

```python
## 231. Power of Two - LEETCODE

def isPowerOfTwo(n):
    
    if (n<=0): return False
    if (n==1):
        return True 
    while n%2==0:
        n//=2
        
    if n == 1:
        return True
    else:
        return False
    
print(isPowerOfTwo(78))


def isPowerOfTwoRecursive(n):
    
    if (n<=0): return False
    if (n==1):
        return True
    if (n%2!=0):
        return False
    
    return isPowerOfTwoRecursive(n//2)
    
print(isPowerOfTwoRecursive(32))
```

```python
## GCD of Two Numbers - Euclidean Number
# The Euclidean algorithm is a method for efficiently calculating the greatest common divisor (GCD) of two integers

def gcd(a, b):
    if b==0:
        return a
    
    return gcd(b, a%b)

def lcm(a, b):
    return a*b//gcd(a, b) # / - float divison, // - Integer Divison

print(gcd(15, 50))
print(lcm(15, 50))
```

```python

# 50. Pow(x, n) - LEETCODE

class Solution:
    def findPow(self, x,n):
        #base case
        if n==0:
            return 1
        a = self.findPow(x, n//2)
        if n%2==0:
            return a*a
        else:
            return a*a*x

    def myPow(self, x: float, n: int) -> float:
        if n>=0:
            return self.findPow(x, n)
        else:
            return 1/self.findPow(x, n*(-1))
        
print(Solution().myPow(2.0, 10))
```
