## Practice in Time and Space Complexity

### Asymptotic Notation


- Best Case (ω)
- Average Case (θ)
- Worst Case (O)

```python
n = 10
x = 3

for i in range(n):
    if i == x:
        break
    print(i, end=" ")
    
x = n
for i in range(n):
    if i == x:
        break
    print(i, end=" ")
        
    
    
# best case -> 1 operation ->(ω(1))
# worst case -> n operation -> (O(n/2)
# average case -> n/2 operations -> (θ(n))
```

```python
# Linear Time Complexity - O(n)

n = 10
for i in range(n):
    print(i, end=" ")
print("\n")

for i in range(n // 2):
    print(float(i), end=" ")
print("\n")
   
for i in range(n + 10):
    print(i, end=" ")
print("\n")    
m = 10
for i in range(n+m):
    print(i, end=" ")
print("\n")    
```

```python
# Constant Time - O(1)

for i in range(50):
    print(i, end=" ")
```

```python
# Quadratic Time Complexity

n = 10
for i in range(n**2):
    print(i, end=" ")
print("\n")        
for i in range(n):
    for j in range(n):
        print(i,j, end=" ")  
```

```python
#Logarithmic Time (O(log N))

n = 100

while n>0:
    print(n, end=" ")
    n //=2
print("\n")  

    
print("\n")  

i = 1
n = 100

while i <=n:
    print(i, end=" ")
    i*=2
    
```

```python
# linearithmetic Time (O(nlogn)), 
for i in range(n):
    while i <=n:
        print(i, end=" ")
        i*=2
        
    
# Cubic Time (O(n^3)),
while i <=n:
    print(i, end=" ")
    i*=3
    
    
# Factoiral Time (O(n!))
```

### Comparison of Time Complexity

- O(1) < O(log n) < O(n) < O(nlogn) < O(n^2) < O(n^3) < O(2^n) < O(n!)
