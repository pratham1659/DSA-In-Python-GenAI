# Tupple in Python

```python
# Mutable -- Element Can Change
# Immutable --  Element can`t be change`

# A tuple is a collection similar to a Python list. The primary difference is that we cannot modify a tuple once it is created.
# Ordered - They maintain the order of elements.
# Immutable - They cannot be changed after creation.
# Allow duplicates - They can contain duplicate values
# 

# Tuple is Immutable
tup = (1, 2, 3, 4, 5, 6)
print(tup)
print(type(tup))

# tup[0] = 3 # Not allow to change the element of tuple

# List is Mutable
list1 = [1, 2, 3, 4, 5, 6]
list1[0] = 3
print(list1)

```

# Only Supported operations in tupple

- len()
- Negative Indexing
- clear()
- Slicing
- count()
- index()



```python
# tupple is immutable so not possible to edit
tp = (1, 2, 3, 4, 5, 6)
tp[1] = 4
print(tp)
```

```python
## check type in tupple
tp = (1)

print(tp)
print(type(tp))

tupF  = (1.0)
print(tupF)
print(type(tupF))

tupS = ('Hello')
print(tupS)
print(type(tupS))   

tup = (1,)
print(tup)
print(type(tup))    
```

```python
## slicing in tupple

tup = (1,2,3,4,5,6)
print(tup)
print(type(tup))
print(tup[1:3])

```

```python
#Index method in Tupple
tup = (1, 2, 5, 6, 1, 2)
print("Index of 5 in tup:", tup.index(5))
```

```python
# Swapping In Python Traditional Method
a = 22
b = 33

temp = a
a = b
b = temp
print(a,b)
```

```python
# Swapping In Python Traditional Method
a = "Two"
b = "One"


a, b = b, a
print(a, b)
```
