```python
class AtmMachine:
    
    def __init__(self):
        self.pin = ""
        self.balance = 0
        self.menu()
    
    def menu(self):
        while True:
            choice = input(
                """
            Please choose any option
            1. Press 0 to add balance
            2. Press 1 to create pin
            3. Press 2 to change pin
            4. Press 3 to check balance
            5. Press 4 to withdraw
            6. Press any other key to exit
                """
                )
            
            if choice == "0":
                self.add_balance()
            elif choice == "1":
                self.create_pin()
            elif choice == "2":
                self.change_pin()
            elif choice == "3":
                self.check_balance()
            elif choice == "4":
                self.withdraw()
                pass
            else:
                print("Thank you for using the ATM. Goodbye!")
                break 
        
        
    def create_pin(self):
        self.user_pin = input("Enter your pin: ")
        self.pin = self.user_pin
        print("PIN created successfully!")
        
    def add_balance(self):
        amount = int(input("Enter the amount you want to add: "))
        self.balance += amount
        print(f"{amount} is added to your account")
        print("Current Balance: ", self.balance)
    
    def check_balance(self):
        print("Your Current balance is:", self.balance)
    
    def change_pin(self):
        old_pin = input("Enter your current pin: ")
        if old_pin == self.pin:
            self.pin = input("Enter your new pin: ")
            print("PIN changed successfully!")
        else:
            print("Incorrect PIN!")
            self.change_pin()
        
    def withdraw(self):
        amount = int(input("Enter amount to withdraw: "))
        if amount <= self.balance:
            self.balance -= amount
            print(f"{amount} withdrawn successfully.")
        else:
            print("Insufficient balance!")
        
```
