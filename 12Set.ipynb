# Set In Python

A set is a collection of unique data, meaning that elements within a set cannot be duplicated.

For instance, if we need to store information about student IDs, a set is suitable since student IDs cannot have duplicates.

```python
set1 = {2, 3, 3, 4, "hello", "world", 3, 5, "hello", "world"}
print(set1)
print(type(set1))
print(len(set1))
```

```python
list1 = [2, 3, 3, 4, 2, "hello", "world", "live", 55, 66, "hello", "live"]
print(list1)
print(len(list1))

## count number of unique elements in the list
set1 = set(list1)
print(set1)
print(type(set1))
print(len(set1))
```

```python
list1 = [2, 3, 3, 4, 2, "hello", "world", "live", 55, 66, "hello", "live"]
print(list1)

set1 = set(list1)
print(set1)
print(len(set1))

set1.add(1)
set1.add(2)
print(set1)
print(len(set1))
```

## Set Operations

```python
list1 = [2, 3, 3, 4, 2, "hello", "world", "live", 55, 66, "hello", "live"]
print(list1)

set1 = set(list1)
print(set1)
print(len(set1))

set1.add(1) ## add operation
set1.add(2)
print(set1)
print(len(set1))

set1.discard("hello") ## discard operation
set1.remove(2) ## Remove operation

set1.discard("hiii") ## element which is not availble will not give error in discard
# set1.remove("hii") ## element which is not available will give error in remove
print(set1)
print(len(set1))
```

```python
set1 = {2, 3, 3, 4, 2, "hello", "world", "live", 55, 66, "hello", "live"}

if 2 in set1:
    print(True)
else:
    print(False)
    
print(set1)  
set1.clear()

print(set1)

## Note : Indexing not available in set because elements were not stored in order
```

```python
## Copy in set
set1 = {2, 3, 4}
set2 = set1 ## This is deep copy

set3 = set1.copy() ## This is shallow copy

set1.add(5)
print(set1)
print(set2)
print(set3)
```

```python
set1 = {1, 8, 5, 2, 4, 6}
set2 = {8, 2, 9, 6}
set3 = { 3, 6, 7, 8, 3, 2, 9}

## print union
print(set1 | set2)
print(set1.union(set3))


## print Intersection
print(set1 & set2)
print(set1.intersection(set3))


## print difference
print(set1 - set2)
print(set1.difference(set3))

## print symemetric difference
print(set1 ^ set3)
print(set1.symmetric_difference(set3))
```

# Set methods

* add()	Adds an element to the set
* clear()	Removes all the elements from the set
* copy()	Returns a copy of the set
* difference()	Returns a set containing the difference between two or more sets
* difference_update()	Removes the items in this set that are also included in another, specified set
* discard()	Remove the specified item
* intersection()	Returns a set, that is the intersection of two other sets
* intersection_update()	Removes the items in this set that are not present in other, specified set(s)
* isdisjoint()	Returns whether two sets have a intersection or not
* issubset()	Returns whether another set contains this set or not
* issuperset()	Returns whether this set contains another set or not
* pop()	Removes an element from the set
* remove()	Removes the specified element
* symmetric_difference()	Returns a set with the symmetric differences of two sets
* symmetric_difference_update()	inserts the symmetric differences from this set and another
* union()	Return a set containing the union of sets
* update()	Update the set with the union of this set and others
