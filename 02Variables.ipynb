### TypeCasting In Python

```python
var1 = "1"
var2 = "2"

print(var1 + var2)
print(int(var1) + int(var2))
```

#### Types of Typecasting Functions

1. Implicit Typecasting - Perform by Python
2. Explicit Typecasting - Perform by User


```python
#Example of Explicit Typecasting

string = "15"
number = 7

str_num = int(string) #throws an ErrorException if String is not a number.
sum = number + str_num
print("The Sum of both value is:", sum)

```

```python
#Example of Implicit Typecasting

var1 = 7
var2 = 3.0

ans = var1 + var2
print(ans)  #float + float
print(type(ans))
```

## Calculator Using Python

```python
a = 50
b = 34

print("The value of ", a, "+", b, "is: ", a + b)
print("The value of ", a, "-", b, "is: ", a - b)
print("The value of ", a, "*", b, "is: ", a * b)
print("The value of ", a, "/", b, "is: ", a / b)
```
